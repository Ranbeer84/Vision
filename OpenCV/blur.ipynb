```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/giraffe-Kenya.png')

image_rgb = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
kernal_s = 10

average_blur = cv2.blur(image_grey, (kernal_s, kernal_s))

plt.imshow(average_blur, cmap='grey')
```

```python
kernal_s = 31

guassian_blur = cv2.GaussianBlur(image_grey, (kernal_s, kernal_s), 0)

plt.imshow(guassian_blur, cmap='grey')
```

```python
median_blur = cv2.medianBlur(image_grey, 11)

plt.imshow(guassian_blur, cmap='grey')
```

```python
plt.figure(figsize=(14, 12))

plt.subplot(2,2,1)
plt.title("original greyscale")
plt.imshow(image_grey, cmap='grey')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Average Blur")
plt.imshow(average_blur, cmap='grey')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Guassian Blur")
plt.imshow(guassian_blur, cmap='grey')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Median Blur")
plt.imshow(median_blur, cmap='grey')
plt.axis('off')
```
